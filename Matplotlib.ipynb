**Scatter Plot**

all others are like bar plot expext these

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,40,23,18,59,18,23,14,22,28,42,34,46,13,35,38,46,51,59,29,14,48]
y=[1,14,22,28,42,34,46,13,35,38,46,51,59,29,14,4,1,2,3,4,40,23,18,11,33]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[8,4,1,6]
y=[3,4,5,6]
c=["green","red","blue","black"]
sizes=[20,30,40,50]
plt.scatter(x,y,color=c,s=sizes,marker="*")
plt.xlabel("day",fontsize=15)
plt.ylabel("Night",fontsize=12)
plt.title("Meta",fontsize=14)
plt.colorbar()
plt.show()
```

**Bar Plot**

```python
import matplotlib.pyplot as plt
y = [44,32,23,45,54]
x = ["c++","c","c#","java","Python"]
plt.bar(x ,y)
plt.ylabel("Demand")
plt.xlabel("Language")
plt.title("CMK", fontsize=13)
plt.show()
```

```python
plt.bar(x,y,width=0.3,color="y")
plt.show()
```

```python
c=["red","green","blue","yellow","purple"]
plt.bar(x,y,width=0.5,color=c)
plt.show()
```

align = edge make c++ and all other in age we can make them by defalt in center , and edgecolor make the bar color red , while width of linewidth is 5 which show bar outer line as 5 in red . and line style : doted show as , we also have any more styles . ND alpha 0.6 make this fended and dead color .

```python
plt.bar(x,y,width=0.5,color=c,align="edge",edgecolor="red",linewidth=5,linestyle=":", alpha=0.6)
plt.show()
```

its import to put label in bar and legend in new line to show label

```python
plt.bar(x,y,width=0.5,color=c,label="Bilawal")
plt.legend()
plt.show()
```

```python
z = [23,15,14,25,22]
c1=["yellow","blue","pink","green","black"]
c2=["green","black","blue","pink","yellow"]
plt.bar(x,y,width=0.5,label="2020",color=c1)
plt.legend()
plt.bar(x,z,width=0.4,label="2000",color=c2)
plt.legend()
plt.show()
```

**Stem Plot**

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[4,5,1,2,3]
plt.stem(x,y)
plt.show()
```

```python
plt.stem(x,y,linefmt=":")
plt.show()
```

markerfmt make it red + we can make it green or anyother color

```python
plt.stem(x,y,linefmt=":",markerfmt="r+")
plt.show()
```

bottom 9 make its ups and down

```python
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=9)
plt.show()
```

basefmt make it green

```python
plt.stem(x,y,linefmt=":",markerfmt="r+",basefmt="g")
plt.show()
```

**Pie Plot**

```python
import matplotlib.pyplot as plt
x=[10,20,40,60]
y=["c","c++","java","python"]
plt.pie(x)
plt.show()
```

```python
plt.pie(x,labels=y)
plt.show()
```

```python
plt.pie(x)
plt.xlabel(y)
plt.show()
```

```python
ex=[0.3,0.0,0.0,0.0]
c=["g","b","r","y"]
plt.pie(x,labels=y,explode=ex,colors=c,shadow=True,counterclock=True,autopct="%0.3f%%")
plt.legend(loc=0)
plt.show()
```

radius make it small and autopct show % on each one

```python
plt.pie(x,labels=y,colors=c,autopct="%0.1f%%",radius=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
x1=[3,4,5,6]
y=["a","b","V","m"]
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w")
plt.show()
```

**Stack Plot**

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
x1=[2,3,4,5]
x2=[1,3,4,5]
x3=[3,4,5,6]
l=["area1","area2","area3"]
plt.stackplot(x, x1, x2, x3,labels=l)
plt.legend()
plt.show()
```
